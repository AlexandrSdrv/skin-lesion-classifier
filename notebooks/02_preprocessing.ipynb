```python
import pandas as pd
import numpy as np
import os
import shutil
from sklearn.model_selection import train_test_split

df = pd.read_csv('../data/HAM10000_metadata.csv')

df['age'] = df['age'].fillna(df['age'].median())

print("Shape:", df.shape)
print("Missing values:", df.isnull().sum().sum())
```

```python
os.makedirs('../data/images', exist_ok=True)

for part in ['HAM10000_images_part_1', 'HAM10000_images_part_2']:
    part_path = f'../data/{part}'
    for img in os.listdir(part_path):
        shutil.copy(os.path.join(part_path, img), '../data/images/')

print("Total images copied:", len(os.listdir('../data/images')))
```

```python
df['image_path'] = df['image_id'].apply(lambda x: f'../data/images/{x}.jpg')

df['exists'] = df['image_path'].apply(os.path.exists)
print("All images found:", df['exists'].all())

train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['dx'])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['dx'])

print("Train size:", len(train_df))
print("Val size:", len(val_df))
print("Test size:", len(test_df))
```

```python
train_df.to_csv('../data/train.csv', index=False)
val_df.to_csv('../data/val.csv', index=False)
test_df.to_csv('../data/test.csv', index=False)

print("Saved")
```
